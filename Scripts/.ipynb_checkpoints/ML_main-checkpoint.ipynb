#### pre-processing

We first drop the few samples with any nan values (mostly found in the rankings). For preprecessing of the featuers, we simply minmax scale the simple features. For samples_output, we take the mean, std, min, max, skew, and kurtosis of the 300 sampled points and then rescale them too. We split the datasets into thee, a training set for training, a validation set for tuning the models (with gridsearches), and a test set for estimating the performance.

```python
# =============================================== DATA PREPARATION ===============================================

# define function to split 2 xr.datasets
# IN  : xr.dataset to split
# OUT : two xr.datasets split randomly based on test_size and random_state
def train_test_split(xarr, test_size, random_state = 42):
    rs = np.random.RandomState(seed = random_state)
    
    N = len(xarr.itemID)
    N_test = int(test_size * N)
    
    test_index = rs.choice(np.arange(N), (N_test,), replace = False)
    train_index = np.delete(np.arange(N), test_index)
    
    xarr_train, xarr_test = xarr.isel(itemID = train_index), xarr.isel(itemID = test_index)

    return xarr_train, xarr_test

# function to extract the labels from xr.dataset
# IN  : xr.dataset 
# OUT : np.array of names of optimizer with lowest ranking
def extract_labels(xarr):
    return xarr.coords['optimizer'].values[np.argmin(xarr.ranking.values,axis=1)].reshape(-1,)

# function to extract the rankings from xr.dataset
# IN  : xr.dataset 
# OUT : np.array of names of optimizer with lowest ranking
def extract_rankings(xarr):
    return xarr.ranking.values.reshape((-1,5))

# extract multilabels for multilabel classification (SLOW!)
# IN  : xr.dataset containing rankings
# OUT : np.array in shape [n_samples, n_optimizers] with binary label per optimizer
def extract_multilabels(xarr, f):
    rankings = extract_rankings(xarr)
    return np.less_equal(rankings, np.repeat(np.min(rankings, axis=1)*f, 5).reshape((-1,5)))*1

# define function to extract features from xr.dataset
# IN  : xr.dataset you need to extract features from
# OUT : nested np.array with per object the features
def extract_features(xarr):
    features = np.array([xarr.dim, xarr.budget, xarr.noise, xarr.convex, xarr.separable, xarr.multimodal]).T
    
    so = xarr.samples_output.values.reshape((-1,300))
    features_so = np.array([np.mean(so, axis=1), np.std(so, axis=1), 
                     np.min(so, axis=1), np.max(so, axis=1),
                     st.skew(so, axis=1), st.kurtosis(so, axis=1)]).T
    
    return np.concatenate([features, features_so], axis=1)

# =============================================== EVALUTATION ===============================================

# function that integrates single pp with f spanning [1,5]
# IN  : np.array with performance
# OUT : float of area under pps
def integrate_perf_profile(arr):
    return np.trapz(arr, dx=4/len(arr))

# returns table with relevant values for each strategy
# IN  : xarray dataset from compute_performance_profiles method
# OUT : pandas DataFrame with f = 1 and area under curve per strategy
def compare_perf_profiles(xr_f):
    df = {'strategy' : [], 'f_one' : [], 'area' : [], 'score' : []}

    best_pp = xr_f.sel(strategy='best_perf_profile').values.reshape((-1,))
    best_pp_area = integrate_perf_profile(best_pp)
    
    for strategy in xr_f.strategy:
        pp = xr_f.sel(strategy=strategy).values.reshape((-1,))
        pp_area = integrate_perf_profile(pp)
        
        df['strategy'].append(strategy.values)
        df['f_one'].append(pp[0])
        df['area'].append(pp_area)
        df['score'].append(pp[0]/best_pp[0] * pp_area/best_pp_area)

    return pd.DataFrame(df).sort_values('score', ascending=False)
```

```python
%%time 
# remove nan values
dsds = dsds.dropna(dim='itemID', how='any')

# split data and extract labels
dsds_train, dsds_test = train_test_split(dsds, 0.3)
dsds_val, dsds_test = train_test_split(dsds_test, 0.5)

# put data into np arrays for training
X_train = extract_features(dsds_train)
y_train = extract_labels(dsds_train)
r_train = extract_rankings(dsds_train)

# applying minmaxscaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
```

## classification approach

For classification, we take two approaches, multiclass classification (predicting one of five classes) and multi label classification (predicting zero or more of five classes), which considers five binary classification problems. The multilabel labels are determined by first setting a factor f. When a ranking of an optimizer is withing that factor of the best ranking its binary label is 1. For the gridsearch of multilabel classification, we use a custom scorer that considers the prediction accurate if at least one of the correct optimizers is predicted.

#### custom scorer

```python
# define custom scorer function for gridsearch
# the fraction of predicted samples that predict at least one correct optimizer
# but no incorrect optimizer
def multilabel_scorer_func(ym_true, ym_pred):    
    union = (np.sum(ym_true*ym_pred, axis=1) > 0)
    return np.less_equal(np.sum(ym_pred, axis=1), union).sum()/len(ym_true)

multilabel_scorer = make_scorer(multilabel_scorer_func)
```

## regression approach

For the regression approach, we approach it as a multi_output regression where the targets are the rankings. We did not apply any preprocessing to the rankings (meaning they have very different orders of magnitude!). For the gridsearch, we define a custom scorer function that calculates the accuracy based on if the model predicts the lowest ranking for the right optimizer.

### gridsearches for regression

```python
# define custom scorer function for gridsearch
def regression_scorer_func(r_true, r_pred):
    return np.sum(np.argmin(r_pred, axis=1) == np.argmin(r_true, axis=1))/len(r_true)

regression_scorer = make_scorer(regression_scorer_func)
```
