## notebook variables

```python
# project directory
PROJECT_DIR = 'L2O_project_datasets/to_SURFFILE'

# notebook constants
TEST_SIZE = 0.15
VAL_SIZE = 0.15
STRATIFIED = True
F = 1.25
RANDOM_STATE = 42
```

## data preparation

#### loading dataset

```python
big_dataset = ExperimentData.from_file(PROJECT_DIR)

%time dsds = open_all_datasets_post(big_dataset)
```

#### pre-processing

We first drop the few samples with any nan values (mostly found in the rankings). For preprecessing of the featuers, we simply minmax scale the simple features. For samples_output, we take the mean, std, min, max, skew, and kurtosis of the 300 sampled points (30 points per 10 realizations) and then rescale them too. We split the datasets into thee, a training set for training, a validation set for tuning the models (with gridsearches), and a test set for estimating the performance.

```python
%%time 
# remove samples with nan values
dsds = dsds.dropna(dim='itemID', how='any')

# split data
dsds_train, dsds_test = train_test_split(dsds, test_size=VAL_SIZE+TEST_SIZE, 
                                         stratified=STRATIFIED, random_state=RANDOM_STATE)
dsds_val, dsds_test = train_test_split(dsds_test, test_size=TEST_SIZE/(VAL_SIZE+TEST_SIZE), 
                                       stratified=STRATIFIED, random_state=RANDOM_STATE)

# put data into np arrays for training
X_train, X_val, X_test = extract_features(dsds_train), extract_features(dsds_val), extract_features(dsds_test)
y_train, y_val, y_test = extract_labels(dsds_train), extract_labels(dsds_val), extract_labels(dsds_test)
#ym_train, ym_val, ym_test = extract_multilabels(dsds_train, F), extract_multilabels(dsds_val, F), extract_multilabels(dsds_test, F)
#r_train, r_val, r_test = extract_rankings(dsds_train), extract_rankings(dsds_val), extract_rankings(dsds_test)
```

```python
# feature processing
pipe = Pipeline((
        ['poly', PolynomialFeatures(degree=3, include_bias=False)], 
        ['selector', SelectFromModel(DecisionTreeClassifier(random_state=RANDOM_STATE), threshold='mean')], 
        ['scaler',  MinMaxScaler()],
        ['clf', FunctionTransformer()]
    ))
    
    
Xt_train = pipe.fit_transform(X_train, y_train)
Xt_val = pipe.transform(X_val)
```

```python
Xt_train.shape
```

## classification approach

For classification, we take two approaches, multiclass classification (predicting one of five classes) and multi label classification (predicting zero or more of five classes), which considers five binary classification problems. The multilabel labels are determined by first setting a factor f. When a ranking of an optimizer is withing that factor of the best ranking its binary label is 1. For the gridsearch of multilabel classification, we use a custom scorer that considers the prediction accurate if at least one of the correct optimizers is predicted.

### untuned, single label

#### training models and calling strategymanager

```python
%%time
# call and train models
models_cls = [
    KNeighborsClassifier(),
    SGDClassifier(random_state = RANDOM_STATE),
    LinearSVC(dual = 'auto', random_state = RANDOM_STATE),
    DecisionTreeClassifier(random_state = RANDOM_STATE),
    RandomForestClassifier(n_jobs = -1, random_state = RANDOM_STATE),
    ExtraTreesClassifier(n_jobs = -1, random_state = RANDOM_STATE),
    AdaBoostClassifier(random_state = RANDOM_STATE)
]

for model in models_cls:
    model.fit(Xt_train, y_train)
    
strategies_cls = [SingleLabelStrategy(model, pipe) for model in models_cls]

strategy_manager_cls = StrategyManager(dsds_test, strategies_cls)
```

## regression approach

For the regression approach, we approach it as a multi_output regression where the targets are the rankings. We did not apply any preprocessing to the rankings (meaning they have very different orders of magnitude!). For the gridsearch, we define a custom scorer function that calculates the accuracy based on if the model predicts the lowest ranking for the right optimizer.

#### training models and calling strategymanager
